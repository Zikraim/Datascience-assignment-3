```python
df = pd.read_csv("train.csv")
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
# Fill missing Age with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill Embarked with mode
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop Cabin column
df = df.drop(columns=["Cabin"], errors="ignore")

# Remove duplicates
df = df.drop_duplicates()
```

```python
survival_gender = df.groupby("Sex")["Survived"].mean()
print(survival_gender)
```

```python
survival_class=df.groupby("Pclass")["Survived"].mean()
print(survival_class)
```

```python
avg_age_class=df.groupby("Pclass")["Age"].mean()
print(avg_age_class)
```

```python
survival_age_group = df.groupby(
    pd.cut(df["Age"], bins=[0, 12, 18, 60, 120],
           labels=["Child", "Teen", "Adult", "Senior"])
)["Survived"].mean()

print(survival_age_group)
```

```python
female_survived = df[(df["Sex"] == "female") & (df["Survived"] == 1)]
print(female_survived)
```

```python
children_survived = df[(df["Age"] < 12) & (df["Survived"] == 1)]
print(children_survived)
```

```python
first_class_survived = df[(df["Pclass"] == 1) & (df["Survived"] == 1)]
print(first_class_survived)
```
